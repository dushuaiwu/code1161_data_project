```python
import xlrd
import matplotlib
from collections import Counter
from pyecharts.charts import Pie
from matplotlib import pyplot as plt
from pyecharts import options as opts
from pyecharts.charts import Map
import pyecharts.options as opts
from pyecharts.charts import Line
import os
```

```python
def data_load():
    # open xlrd
    data = xlrd.open_workbook('./data.xlsx')
    table = data.sheets()[0]
    nrows = table.nrows  # Ë°å
    ncols = table.ncols  # Âàó
    all_data = []

    # Loading data
    for i in range(0, nrows):
        data = []
        rowValues = table.row_values(i)  # 
        for item in rowValues:
            data.append(item)
        all_data.append(data)


    country = []

    country_name = []

    population = []

    infections_number = []

    death_number = []

    total_vaccinations = []

    people_fully_vaccinated = []

    vaccine = []

    for i in range(len(all_data)):

        country_n = all_data[i][5]
        country.append(country_n)

        name = all_data[i][4]
        country_name.append(name)

        popu = all_data[i][0]
        population.append(popu)

        infect = all_data[i][2]
        infections_number.append(infect)

        death = all_data[i][3]
        death_number.append(death)

        vaccinations = all_data[i][7]
        total_vaccinations.append(vaccinations)

        fully_vaccinated = all_data[i][9]
        people_fully_vaccinated.append(fully_vaccinated)

        vac = all_data[i][14]
        vaccine.append(vac)

    country = country[1:]
    population = population[1:]
    infections_number = infections_number[1:]
    death_number = death_number[1:]
    total_vaccinations = total_vaccinations[1:]
    people_fully_vaccinated = people_fully_vaccinated[1:]
    vaccine = vaccine[1:]
    country_name = country_name[1:]

    # print(country)
    # print(population)
    # print(infections_number)
    # print(death_number)
    # print(total_vaccinations)
    # print(people_fully_vaccinated)
    # print(vaccine)
    # print(country_name)
    return country, population, infections_number, death_number, total_vaccinations, people_fully_vaccinated, vaccine, country_name
```

# 2.Infectionüßü‚Äç‚ôÇÔ∏èüßü‚Äç‚ôÄÔ∏èüßü‚Äç‚ôÄÔ∏è


```python
# covid_19 Infection rate
def each_country_infections_rate(country, infections_number, population):
    infections_rate = []
    for i in range(len(country)):
        rate = float('{:.2}'.format(infections_number[i] / population[i]))
        infections_rate.append(rate)
    print(infections_rate)
    plt.rcParams['font.sans-serif'] = ['SimHei']
    matplotlib.rc('font', family='SimHei', weight='bold')
    plt.barh(range(len(infections_rate)), infections_rate, tick_label=country)
    plt.title('covid_19 Infection rate')
    plt.show()
```

```python
each_country_infections_rate(country, infections_number, population)
```

# 4.Vaccination üß™üß™üß™

```python
def each_country_Vaccination_rate(country, total_vaccinations, population):
# Vaccination rates in each countries
    Vaccination_rate = []
    for i in range(len(country)):
        rate = float('{:.2}'.format(total_vaccinations[i]/population[i]))
        Vaccination_rate.append(rate)
    print(Vaccination_rate)
    plt.bar(country, Vaccination_rate)
    plt.title('Vaccination rates in each countries')
    plt.show()
```

```python
each_country_Vaccination_rate(country, total_vaccinations, population)
```

```python
def each_country_fully_Vaccination_rate(country, people_fully_vaccinated, population):
# Fully vaccination rate in each country
    fully_Vaccination_rate = []
    for i in range(len(country)):
        rate = float('{:.2}'.format(people_fully_vaccinated[i]/population[i]))
        fully_Vaccination_rate.append(rate)
    plt.pie(fully_Vaccination_rate, labels=country, wedgeprops=dict(width=0.3, edgecolor='w'))
    plt.axis('equal')
    plt.show()
```

```python
each_country_fully_Vaccination_rate(country, people_fully_vaccinated, population)
```

```python
def each_country_fully_Vaccination_rate2(country, people_fully_vaccinated, total_vaccinations):
# Full coverage rate of vaccinated people
    fully_Vaccination_rate2 = []
    for i in range(len(country)):
        rate = float('{:.2}'.format(people_fully_vaccinated[i]/total_vaccinations[i]))
        fully_Vaccination_rate2.append(rate)
    print(fully_Vaccination_rate2)
    plt.title('Full coverage rate of vaccinated people')
    plt.plot(country,fully_Vaccination_rate2,'s-',color = 'r',label="ATT-RLSTM")#s-:sqare
    plt.show()
```

```python
each_country_fully_Vaccination_rate2(country, people_fully_vaccinated, total_vaccinations)
```

```python
def each_country_vaccine_type(vaccine):
# different vaccines
    vaccine_type = []
    for va in vaccine:
        if '/' in va:
            va = str(va)
            type = va.split('/')
            for i in type:
                vaccine_type.append(i)
        else:
            vaccine_type.append(va)

    vaccine_type_number = dict(Counter(vaccine_type))
    
    vaccine_name = list(vaccine_type_number)
    vaccine_number = list(vaccine_type_number.values())
    patches,text1,text2 = plt.pie(vaccine_number,
                          labels=vaccine_name,
                          labeldistance = 1.2,
                          autopct = '%3.2f%%',
                          shadow = False,
                          startangle =90,
                          pctdistance = 0.6)
    plt.axis('equal')
    plt.legend()
    plt.show()
```

```python
each_country_vaccine_type(vaccine)
```
